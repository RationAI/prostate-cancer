## Per-Model Non-Inferiority

```python
import numpy as np
from numpy.typing import NDArray
from sklearn.metrics import roc_auc_score

def bootstrap_auc_diff(
    y: NDArray[np.int_],
    scores_max: NDArray[np.floating],
    scores_other: NDArray[np.floating],
    n_boot: int = 10_000,
    seed: int = 0
) -> NDArray[np.floating]:
    """
    Bootstrap distribution of AUC differences:

        Δ = AUC(other) - AUC(max)

    Parameters
    ----------
    y : array-like of shape (n_samples,)
        Binary ground-truth labels (0/1).
    scores_max : array-like of shape (n_samples,)
        Prediction scores from max aggregation.
    scores_other : array-like of shape (n_samples,)
        Prediction scores from another aggregation method.
    n_boot : int
        Number of bootstrap resamples.
    seed : int
        Random seed for reproducibility.

    Returns
    -------
    NDArray[np.float_]
        Bootstrap samples of ΔAUC.
    """
    """
    Returns bootstrap distribution of:
        Δ = AUC(other) - AUC(max)
    """
    rng = np.random.default_rng(seed)
    n = len(y)

    deltas = []

    for _ in range(n_boot):
        idx = rng.integers(0, n, n)

        y_b = y[idx]

        # skip degenerate samples
        if len(np.unique(y_b)) < 2:
            continue

        auc_max = roc_auc_score(y_b, scores_max[idx])
        auc_other = roc_auc_score(y_b, scores_other[idx])

        deltas.append(auc_other - auc_max)

    return np.array(deltas)
```

```python
from typing import Any

def non_inferiority_summary(
    deltas: np.ndarray,
    delta: float,
) -> dict[str, Any]:
    """
    Convert bootstrap Δ distribution into NI decision
    """

    mean_delta = float(np.mean(deltas))
    lower = float(np.percentile(deltas, 2.5))
    upper = float(np.percentile(deltas, 97.5))
    upper_95 = float(np.percentile(deltas, 95))

    return {
        "mean_delta": mean_delta,
        "ci_95_lower": lower,
        "ci_95_upper": upper,
        "upper_95": upper_95,
        "non_inferior": upper_95 < delta
    }


def evaluate_single_model(
    model_name: str,
    y: np.ndarray,
    delta: float = 0.01,
    n_boot: int = 10_000
) -> pd.DataFrame:
    """
    Compare max vs all aggregation methods for one model
    """

    results = []

    preds = model_preds[model_name]
    max_scores = np.asarray(preds["max"])

    for method in ["mpm", "topk", "span"]:

        other_scores = np.asarray(preds[method])

        deltas = bootstrap_auc_diff(
            y=y,
            scores_max=max_scores,
            scores_other=other_scores,
            n_boot=n_boot,
            seed=42
        )

        stats = non_inferiority_summary(deltas, delta=delta)

        stats.update({
            "model": model_name,
            "method": method
        })

        results.append(stats)

    return pd.DataFrame(results)
```

```python
margin = 0.02 # We consider max acceptable if it is not worse by more than 0.02 AUC

pgp_df = evaluate_single_model("pgp", gt, margin)
virchow2_df = evaluate_single_model("virchow2", gt, margin)
resnet50_df = evaluate_single_model("resnet50", gt, margin)
resnet50_stain_df = evaluate_single_model("resnet50_stain", gt, margin)
vgg16_df = evaluate_single_model("vgg16", gt, margin)
vgg16_stain_df = evaluate_single_model("vgg16_stain", gt, margin)
gvit_df = evaluate_single_model("gvit", gt, margin)
gvit_stain_df = evaluate_single_model("gvit_stain", gt, margin)
```
