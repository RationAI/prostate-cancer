## Master DF

```python
import mlflow
import pandas as pd


master_df = None


def fetch_table(uri: str) -> str:
    return mlflow.artifacts.download_artifacts(uri)


assert len(DATA) == 8  # 8 models

for model_name, methods in DATA.items():
    assert len(methods) == 4  # 4 aggr. methods

    for pred_type, (uri, pred_col) in methods.items():
        assert pred_type in ["max", "mpm", "topk", "span"]

        df = pd.read_json(fetch_table(uri), orient="split")

        cols = ["slide_name", pred_col]
        if master_df is None:
            cols.append("target")

        tmp = df[cols].rename(columns={pred_col: f"{model_name}__{pred_type}"})

        if master_df is None:
            master_df = tmp
        else:
            master_df = master_df.merge(
                tmp,
                on="slide_name",
                how="inner",
                validate="one_to_one",
            )

assert len(master_df) == 424
```

## Per-Model Non-Inferiority

```python
import numpy as np
from numpy.typing import NDArray
from sklearn.metrics import roc_auc_score


def bootstrap_auc_diff(
    y: NDArray[np.int_],
    scores_max: NDArray[np.floating],
    scores_other: NDArray[np.floating],
    n_boot: int = 10_000,
    seed: int = 0,
) -> NDArray[np.floating]:
    """Bootstrap distribution of AUC differences.

        Δ = AUC(other) - AUC(max)

    Parameters
    ----------
    y : array-like of shape (n_samples,)
        Binary ground-truth labels (0/1).
    scores_max : array-like of shape (n_samples,)
        Prediction scores from max aggregation.
    scores_other : array-like of shape (n_samples,)
        Prediction scores from another aggregation method.
    n_boot : int
        Number of bootstrap resamples.
    seed : int
        Random seed for reproducibility.

    Returns:
    -------
    Bootstrap distribution of Δ = AUC(other) - AUC(max).
    """
    rng = np.random.default_rng(seed)
    n = len(y)

    deltas = []

    for _ in range(n_boot):
        idx = rng.integers(0, n, n)

        y_b = y[idx]

        # skip degenerate samples
        if len(np.unique(y_b)) < 2:
            continue

        auc_max = roc_auc_score(y_b, scores_max[idx])
        auc_other = roc_auc_score(y_b, scores_other[idx])

        deltas.append(auc_other - auc_max)

    return np.array(deltas)
```

```python
from typing import Any


def non_inferiority_summary(
    deltas: np.ndarray,
    delta: float,
) -> dict[str, Any]:
    """Convert bootstrap Δ distribution into NI decision."""
    mean_delta = float(np.mean(deltas))
    lower = float(np.percentile(deltas, 2.5))
    upper = float(np.percentile(deltas, 97.5))
    upper_95 = float(np.percentile(deltas, 95))

    return {
        "mean_delta": mean_delta,
        "ci_95_lower": lower,
        "ci_95_upper": upper,
        "upper_95": upper_95,
        "non_inferior": upper_95 < delta,
    }


def evaluate_single_model(
    master_df: pd.DataFrame, model_name: str, delta: float = 0.01, n_boot: int = 10_000
) -> pd.DataFrame:
    """Compare max vs all aggregation methods for one model."""
    results = []

    max_preds = np.asarray(master_df[f"{model_name}__max"])
    gt = np.asarray(master_df["target"])

    for method in ["mpm", "topk", "span"]:
        other_preds = np.asarray(master_df[f"{model_name}__{method}"])

        deltas = bootstrap_auc_diff(
            y=gt,
            scores_max=max_preds,
            scores_other=other_preds,
            n_boot=n_boot,
            seed=42,
        )

        stats = non_inferiority_summary(deltas, delta=delta)

        stats.update({"model": model_name, "method": method})

        results.append(stats)

    return pd.DataFrame(results)
```

```python
margin = 0.02  # We consider max acceptable if it is not worse by more than 0.02 AUC

pgp_df = evaluate_single_model(master_df, "pgp", margin)
virchow2_df = evaluate_single_model(master_df, "virchow2", margin)
resnet50_df = evaluate_single_model(master_df, "resnet50", margin)
resnet50_stain_df = evaluate_single_model(master_df, "resnet50_stain", margin)
vgg16_df = evaluate_single_model(master_df, "vgg16", margin)
vgg16_stain_df = evaluate_single_model(master_df, "vgg16_stain", margin)
gvit_df = evaluate_single_model(master_df, "gvit", margin)
gvit_stain_df = evaluate_single_model(master_df, "gvit_stain", margin)
```
